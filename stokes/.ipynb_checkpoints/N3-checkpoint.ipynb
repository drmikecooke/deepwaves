```python
import sympy as sm
from sympy.simplify.fu import TR8
```

```python
b,p=sm.symbols('b \\phi')
A_1=b

A_22=sm.Symbol('A_{22}')
A_2=A_22*b**2
```

```python
x=-p-A_1*sm.sin(p)-A_2*sm.sin(2*p)
x
```

```python
y=A_1*sm.cos(p)+A_2*sm.cos(2*p)
y
```

```python
F_2,K_2=sm.symbols('F_2 K_2')
F=1+F_2*b**2
K=K_2*b**2
```

```python
S=x.diff(p)**2+y.diff(p)**2
S
```

```python
series=((F+K-2*y)*S-F).series(b,n=3)
TR8(series)
```

```python
import numpy as np
from matplotlib.pyplot import plot
P=np.linspace(0,2*np.pi)
for bi in np.linspace(0,0.43,5):
    X=[x.subs(b,bi).subs(p,pi) for pi in P]
    Y=[y.subs(b,bi).subs(p,pi) for pi in P]
    plot(X,Y)
```
