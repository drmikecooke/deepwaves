```python
import sympy as sm
from sympy.simplify.fu import TR8
```

```python
b,p=sm.symbols('b \\phi')
A_1=b

A_33=sm.Symbol('A_{33}')
A_2=b**2
A_3=A_33*b**3
```

```python
x=-p-A_1*sm.sin(p)-A_2*sm.sin(2*p)-A_3*sm.sin(3*p)
x
```

```python
y=A_1*sm.cos(p)+A_2*sm.cos(2*p)+A_3*sm.cos(3*p)
y
```

```python
F_2=sm.symbols('F_2')
F=1+F_2*b**2
K=b**2
```

```python
S=x.diff(p)**2+y.diff(p)**2
S
```

```python
series=((F+K-2*y)*S-F).series(b,n=4)
TR8(series)
```

```python
x=x.subs(A_33,sm.Rational(3,2))
x
```

```python
y=y.subs(A_33,sm.Rational(3,2))
x
```

```python
import numpy as np
from matplotlib.pyplot import plot
P=np.linspace(0,2*np.pi)
for bi in np.linspace(0,0.43,5):
    X=[x.subs(b,bi).subs(p,pi) for pi in P]
    Y=[y.subs(b,bi).subs(p,pi) for pi in P]
    plot(X,Y)
```
