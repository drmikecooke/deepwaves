```python
import sympy as sm
from sympy.simplify.fu import TR8
```

```python
b,p=sm.symbols('b \\phi')
A_1=b

A_35,A_55=sm.symbols('A_{35} A_{55}')
A_2=b**2+sm.Rational(1,2)*b**4
A_3=sm.Rational(3,2)*b**3+A_35*b**5
A_4=sm.Rational(8,3)*b**4
A_5=A_55*b**5
```

```python
x=-p-A_1*sm.sin(p)-A_2*sm.sin(2*p)-A_3*sm.sin(3*p)-A_4*sm.sin(4*p)-A_5*sm.sin(5*p)
x
```

```python
y=A_1*sm.cos(p)+A_2*sm.cos(2*p)+A_3*sm.cos(3*p)+A_4*sm.cos(4*p)+A_5*sm.cos(5*p)
y
```

```python
F_4=sm.symbols('F_4')
F=1+b**2+F_4*b**4
K=b**2+2*b**4
```

```python
S=x.diff(p)**2+y.diff(p)**2
S
```

```python
(TR8(((F+K-2*y)*S-F).expand().collect(b))/b**5).expand().subs(b,0)
```

```python
series=((F+K-2*y)*S-F).subs(A_55,sm.Rational(125,24)).subs(A_35,sm.Rational(19,12)).subs(F_4,sm.Rational(7,2)).series(b,n=5)
TR8(series)
```

```python
x=x.subs(A_55,sm.Rational(125,24)).subs(A_35,sm.Rational(19,12)).subs(F_4,sm.Rational(7,2))
x
```

```python
y=y.subs(A_55,sm.Rational(125,24)).subs(A_35,sm.Rational(19,12)).subs(F_4,sm.Rational(7,2))
y
```

```python
import numpy as np
from matplotlib.pyplot import plot
P=np.linspace(0,2*np.pi)
for bi in np.linspace(0,0.2,5):
    X=[x.subs(b,bi).subs(p,pi) for pi in P]
    Y=[y.subs(b,bi).subs(p,pi) for pi in P]
    plot(X,Y)
```
