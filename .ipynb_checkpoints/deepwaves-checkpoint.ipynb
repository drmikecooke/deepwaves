```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import root

from deepwaves import pack,unpack,IEj,linear
```

```python
def F(A,H,d):
    '''Set of conditions on boundaries'''
    Ej,Bj,Ub,Q,R=unpack(A)
    N=len(Ej)
    Etam=IEj([d,*Ej],Xm) # imposes depth constraint
    Etamj=Etam.reshape((N+1,1))@j.reshape((1,N))
    CEmj,SEmj=C(Etamj)/Chjd,S(Etamj)/Chjd
    kin=-Ub*Etam+((SEmj*CXmj))@(Bj)+Q # 2N+1 kinetic conditions
    Um=-Ub+((CEmj*CXmj))@(j*Bj)
    Vm=((SEmj*SXmj))@(j*Bj)
    dyn=(Um**2+Vm**2)/2+Etam-R # 2N dynamic conditions
    height=Etam[0]-Etam[-1] # assume height is maximum-minium given by eta at x=0,pi, respectively
    return np.hstack([kin,dyn,height-H])

class grids:
    def __init__(N):
        self.Xm=np.linspace(0,np.pi,N+1) # N+1 grid for collocation
        self.j=np.arange(1,N+1) # Fourier index from 1 to N
        self.Xmj=Xm.reshape((N+1,1))@j.reshape((1,N)) # create matrix of X-grid multiplied by Fourier index
```

```python
def test(N,H,d):
    global Xm,j,C,S,Chjd,CXmj,SXmj # perhaps would be better to localize these at later stage?
    # I use k,g=1,1 units so wavelength is 2pi
    A0=linear(N,H,d) # use linear approximation to start root finding
    Xm=np.linspace(0,np.pi,N+1) # N+1 grid for collocation
    j=np.arange(1,N+1) # Fourier index from 1 to N
    Xmj=Xm.reshape((N+1,1))@j.reshape((1,N)) # create matrix of X-grid multiplied by Fourier index
    # Chjd is Bj scaling factor - in Fenton it is cosh(jd), but I find cosh(j(H+d)) enables higher waves
    CXmj,SXmj,Chjd=np.cos(Xmj),np.sin(Xmj),np.cosh(j*(H+d)) # evaluate sines and cosines for Fourier expansions, and Chjd
    C,S=(np.exp,np.exp) if d==0 else (np.cosh,np.sinh) # hyperbolics for Y dependence, for deep water use exp
    return root(F,A0,args=(H,d)),H,d # deliver root of F which encodes boundary conditions, and Fenton eta Fourier components (E)
```

```python
# even n
X=np.linspace(0,np.pi,100)
for n in range(2,13,2): # step two at time so graph not too busy
    sol,H,d=test(20,0.02*n*np.pi,0)
    if not sol.success: # weed out fails and print analysis
        print(sol)
        continue
    Ej,Bj,Ub,Q,R=unpack(sol.x)
    plt.plot(X,IEj([d,*Ej],X),label=f"Num: H/L={H/2/np.pi:6.4f}")
plt.legend()
plt.title(f'Test d=0, deep water')
plt.show()
```

```python
# odd n
for n in range(1,15,2): # step two at time so graph not too busy
    sol,H,d=test(10,0.02*n*np.pi,0)
    if not sol.success: # weed out fails and print analysis
        print(sol)
        continue
    Ej,Bj,Ub,Q,R=unpack(sol.x)
    plt.plot(X,IEj([d,*Ej],X),label=f"Num: H/L={H/2/np.pi:6.4f}")
plt.legend()
plt.title(f'Test d=0, deep water')
plt.show()
```
