# Comparison with Fenton's Fourier.exe

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import root

from deepwaves import pack,unpack
```

```python
def get(n): # Read Fenton result n
    with open(f"deep/{n:02d}.txt","rt") as fenton:
        comp=fenton.read()
    
    comp=comp.split('\n')
    
    targets=['d','H','c','Q','R','q','r']
    data={}
    jBE=[]
    for line in comp:
        if line:
            for target in targets:
                if f'({target})' in line:
                    data[target]=float(line.split('\t')[1])
            if '0'<=line[1]<='9':
                jBE+=[line.split('\t')]
            
    data['Ub']=data.pop('c') # Ub is wave speed in my code
    if not 'd' in data:
        data['d']=0 # deep water
    J,B,E=np.array(jBE).T
    return data,J.astype(int),B.astype(float),E.astype(float)
```

```python
def F(A):
    '''Set of conditions on boundaries'''
    Ej,Bj,Ub,Q,R=unpack(A)
    Etam=IEj0([d,*Ej],Xm) # imposes depth constraint
    Etamj=Etam.reshape((N+1,1))@j.reshape((1,N))
    CEmj,SEmj=C(Etamj)/Chjd,S(Etamj)/Chjd
    kin=-Ub*Etam+((SEmj*CXmj))@(Bj)+Q # 2N+1 kinetic conditions
    Um=-Ub+((CEmj*CXmj))@(j*Bj)
    Vm=((SEmj*SXmj))@(j*Bj)
    dyn=(Um**2+Vm**2)/2+Etam-R # 2N dynamic conditions
    height=Etam[0]-Etam[-1] # assume height is maximum-minium given by eta at x=0,pi, respectively
    return np.hstack([kin,dyn,height-H])

def IEj(Ej,X):
    '''Use Ej to interpolate eta at values in X-array'''
    N=len(Ej)
    CXj=np.cos(X.reshape(len(X),1)@np.arange(1,N+1).reshape(1,N))
    return d+CXj@Ej

def IEj0(Ej,X):
    '''Use Ej to interpolate eta at values in X-array, Ej[0]=d'''
    N=len(Ej)
    CXj=np.cos(X.reshape(len(X),1)@np.arange(0,N).reshape(1,N))
    return CXj@Ej
```

```python
def test(n):
    global d,Xm,N,j,C,S,Chjd,CXmj,SXmj,H # perhaps would be better to localize these at later stage?
    data,J,B,E=get(n) # Fenton result n
    N,H,d=20,0.02*n*np.pi,0 # I use k,g=1,1 units so wavelength is 2pi
    Bj0=np.zeros(N) # initial guess based on linear approximation
    Ej0=np.zeros(N)
    Ub0=1 if d==0 else np.tanh(d)**(1/2)
    Bj0[0]=H/2/Ub0
    Ej0[0]=H/2
    Q0,R0=Ub0*d,Ub0**2/2+d
    
    A0=pack(Ej0,Bj0,Ub0,Q0,R0) # pack them into vector

    Xm=np.linspace(0,np.pi,N+1) # N+1 grid for collocation
    j=np.arange(1,N+1) # Fourier index from 1 to N
    Xmj=Xm.reshape((N+1,1))@j.reshape((1,N)) # create matrix of X-grid multiplied by Fourier index
    # Chjd is Bj scaling factor - in Fenton it is cosh(jd), but I find cosh(j(H+d)) enables higher waves
    CXmj,SXmj,Chjd=np.cos(Xmj),np.sin(Xmj),np.cosh(j*(H+d)) # evaluate sines and cosines for Fourier expansions, and Chjd
    C,S=(np.exp,np.exp) if d==0 else (np.cosh,np.sinh) # hyperbolics for Y dependence, for deep water use exp
    return root(F,A0),E # deliver root of F which encodes boundary conditions, and Fenton eta Fourier components (E)
```

```python
X=np.linspace(0,np.pi,100)
for n in range(2,13,2):
    sol,E=test(n)
    if not sol.success: # weed out fails and print analysis
        print(sol)
        continue
    Ej,Bj,Ub,Q,R=unpack(sol.x)
    plt.plot(X,IEj(Ej,X),label=f"Num: H={H:6.4f}")
    plt.plot(X,IEj(E.astype(float),X),label=f"Fenton: H={H:6.4f}")
plt.legend()
plt.title(f'Comparison')
plt.show()
```

```python
X=np.linspace(0,np.pi,200)
# Deviation increases as highest wave approached, reducing upper limit (13=12+1), shows deviation for less high waves
for n in range(2,13,2): 
    sol,E=test(n)
    if not sol.success: # weed out fails and print analysis
        print(sol)
        continue
    Ej,Bj,Ub,Q,R=unpack(sol.x)
    plt.plot(X,IEj(E.astype(float),X)-IEj(Ej,X),label=f"Fenton-Num.: H={H:6.4f}")
plt.legend()
plt.title(f'Difference: Fenton $-$ my code')
plt.show()
```
